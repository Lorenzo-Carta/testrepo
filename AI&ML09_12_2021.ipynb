```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dataset/iris.data")
df.head()
new_row={"sepal_length":5.1,"sepal_width":3.5,"petal_length":1.4,"petal_width":0.2,"class":"Iris-setosa"}
```

```python
df.rename(columns={"5.1":"sepal_length","3.5":"sepal_width","1.4":"petal_length","0.2":"petal_width","Iris-setosa":"class"},inplace=True)
df=df.append(new_row,ignore_index=True)
df.head()
df.shape

```

```python
df_set=df[df["class"]=="Iris-setosa"]
df_set.head()
```

```python
sns.set_style(style="darkgrid")
_=sns.histplot(df_set["petal_width"],bins=5)
plt.show()
sns.regplot(df["sepal_length"],df["petal_length"])
```

```python
sns.set_style(style="darkgrid")
fig,((ax1,ax2,ax5),(ax3,ax4,ax6))=plt.subplots(2,3,figsize=(15,10))

sns.boxplot(x="class",y="sepal_length",data=df,ax=ax1)
sns.boxplot(x="class",y="sepal_width",data=df,ax=ax2)
sns.regplot(x="sepal_length",y="sepal_width",data=df,ax=ax5,color="red")
sns.boxplot(x="class",y="petal_length",data=df,ax=ax3)
sns.boxplot(x="class",y="petal_width",data=df,ax=ax4)
sns.regplot(x="petal_length",y="petal_width",data=df,ax=ax6,color="green")
ax1.set_title("BOXPLOT SEPAL LENGHT")
ax2.set_title("BOXPLOT-SEPAL WIDHT")
ax3.set_title("BOXPLOT PETAL LENGHT")
ax4.set_title("BOXPLOT PETAL WIDTH")
ax5.set_title("SCATTER PLOT SEPAL LENGTH-WIDTH")
ax6.set_title("SCATTER PLOT PETAL LENGTH-WIDTH")
plt.savefig("Fisher EDA")
```
